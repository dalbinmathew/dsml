SVM: support vector machine
is a classification algorithm,supervised learning algorithm that classifies data
it can be used for linear or non linear classification
used for regression problem.
used for outliers detection task
linear is a straight line
no linear is a 3d structure(hyper plane)
main concept is maximum margin
it can have anomalies(outliers)
if outliers occurs, either ignore or misclassify it.
2 errors are used in this 
  1.classification errror(target-actual)
  2.margin error (maximum margin)
  application:
    image classification,sendimental analysis,image recognition,handwritting identification,spam detection,face detection,anomaly detection ,gene expression analysis etc.
SVM LINEAR:
  it is a 2d surface,straight line,that classifies data points.
  it is also called maximum margin linear classifier -setting maximum margin.
  it ensures that the margin between the closest data points of different classes(support vectors)is maximized
SVM NONLINEAR:
  it is a 3d structure(hyper plane)that classifies data points.
  the dimension of the hyper plane depends on the no of features
  in this,there can be multiple planes or discission boundaries to segreggate the classes in n dimensional space - we need to find out the best decision boundary that help tonclassify the data points,this best boundary is known as the hyper plane.simply svm is a machine learning algorithm to find the hyper plane in any n dimentional space, that distinctly classify the data points.
  when outliers are there ,need to misclassify or ignore it.
  if we consider outliers- minimum margin - wrong prediction

soft margin formulation:
  it allows svm to make a certain no of mistakes and keep margin as wide as posssible.so that other data points can still be classified correctly.

how to perform misclassification?
> with the help of hyper parameter called c parameters.it is the svm how much you want to avoid misclassifiying in each training.
> when you increse the value of c parameter - minimum margin
if c=1000 ,minimum margin if c=1 ,maximum margin

ERRORS IN SVM:
  1.classification error (a)
  2.margin error (b)

SVM GOAL = MIN(Ca+b)







```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import numpy as np


```

```python
X,y = make_classification(n_samples=1000,n_features=2,n_informative=2,n_redundant=0,random_state=42)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
s=SVC(kernel='linear').fit(X_train,y_train)
y_pred = s.predict(X_test)
ac=accuracy_score(y_test,y_pred)
print(ac)
report=classification_report(y_test,y_pred)

print(f"classification_report{report}")



```
