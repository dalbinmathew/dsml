```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.datasets import make_classification #used to generate some random data,similar 
import numpy as np
from sklearn.metrics import accuracy_score,classification_report
from matplotlib import pyplot as plt
```

```python
x,y=make_classification(n_features=2,n_samples=1000,random_state=42,n_informative=2,n_redundant=0)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model=SVC(kernel="linear")
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
accuracy_score(y_test,y_predict)
```

```python
report=classification_report(y_test,y_predict)
print(report)
```

```python
#plotting the decision boundary and margin
plt.figure(figsize=(8,6))
```

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap="cividis",s=50,alpha=0.6)
ax=plt.gca()
xlimit=ax.get_xlim()
ylimit=ax.get_ylim()
xx,yy=np.meshgrid(np.linspace(xlimit[0],xlimit[1],100),np.linspace(ylimit[0],ylimit[1],100))
z=model.decision_function(np.c_[xx.ravel(),yy.ravel()])
z=z.reshape(xx.shape)
plt.contour(xx,yy,z,colors="k",levels=[-1,0,1],alpha=0.6,linestyles=["--","-","--"])
plt.title("SVM decision boundary")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
