```python
import matplotlib.pyplot as plt

x=np.arange(1,11)
y=x+5
plt.title("Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y,'x b')
plt.show()
```

```python
x=np.arange(0,3*np.pi,.1)
y=np.sin(x)
z=np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title('sine wave')
plt.subplot(2,1,2)
plt.plot(x,z)
plt.title('cosine wave')
plt.show()
```

```python

x=[5,10,15,20]
y=[45,79,145,234]
z=[]
for i,j in zip(x,y):
    z.append(j/i)
plt.plot(x,z,color='blue',lw=10,label='runrate',)
plt.xlabel('over')
plt.ylabel('runrate')
plt.title("cricket")
plt.show()
```
